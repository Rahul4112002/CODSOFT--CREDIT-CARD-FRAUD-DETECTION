```python
df.drop('Unnamed: 0',axis =1 , inplace=True)
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['hour'] = df['trans_date_trans_time'].dt.hour
df['minute'] = df['trans_date_trans_time'].dt.minute
df['second'] = df['trans_date_trans_time'].dt.second
df['year'] = df['trans_date_trans_time'].dt.year
df['month'] = df['trans_date_trans_time'].dt.month
df['day'] = df['trans_date_trans_time'].dt.day
```

```python
df.drop('trans_date_trans_time',axis =1 , inplace=True)
```

```python
df.iloc[2]
```

# Independent and dependent variable

```python
X = df[['cc_num', 'amt', 'lat', 'long', 'city_pop', 'hour', 'minute', 'second', 'year', 'month', 'day']]
y = df.is_fraud
```

```python
X.info()
```

# Train Test Split

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=42)
```

```python
numerical_cols = ['cc_num', 'amt', 'lat', 'long', 'city_pop', 'hour', 'minute', 'second', 'year', 'month', 'day']
```

```python
#Numerical Pipeline
num_pipeline = Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scalar',StandardScaler())
    ]
)
```

```python
preprocessor = ColumnTransformer(
    [
        ('num_pipeline',num_pipeline,numerical_cols)
    ]
)
preprocessor
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
## Model Training Automation
models={
    'Random Forest':RandomForestClassifier(),
    'Logistic Regression':LogisticRegression(),
    'Decision Tree':DecisionTreeClassifier()
}
```

```python
def evaluate_model(X_train,y_train,X_test,y_test,models):

    report = {}
    for i in range(len(models)):
        model = list(models.values())[i]
        # Train model
        model.fit(X_train,y_train)



        # Predict Testing data
        y_test_pred =model.predict(X_test)

        # Get accuracy for test data prediction

        test_model_score = accuracy_score(y_test,y_test_pred)

        report[list(models.keys())[i]] =  test_model_score



    return report
```

```python
evaluate_model(X_train,y_train,X_test,y_test,models)
```

# Random Forest Classifier

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train , y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print("Classification Report: ", classification_report(y_pred,y_test))
```

```python
print("Confusion Metrix: ", confusion_matrix(y_pred,y_test))
```

```python
print("Accuracy : ",accuracy_score(y_pred,y_test))
```
